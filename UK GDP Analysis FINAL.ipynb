```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
UK_econ_data = 'csv file/19-03-20 12_16_59_theglobaleconomy.csv'
```

```python
UK_econ_df = pd.read_csv(UK_econ_data)
UK_econ_df
```

```python
del UK_econ_df['Retail sales index']
del UK_econ_df['Retail sales Y-on-Y']
del UK_econ_df['Purchasing Managers Index (PMI) services']
del UK_econ_df['Consumption as percent of GDP']
del UK_econ_df['Consumption growth']

UK_econ_df
```

```python
quarterly_data = UK_econ_df.dropna(how='any')
quarterly_data
```

```python
quarters = ['2010', 'Q2', 'Q3', 'Q4', '2011', 'Q2', 'Q3', 'Q4', '2012', 'Q2', 'Q3', 'Q4', '2013',
                    'Q2', 'Q3', 'Q4', '2014', 'Q2', 'Q3', 'Q4']
quarterly_data_copy = quarterly_data.copy()
quarterly_data_copy['Quarters'] = quarters
quarterly_data_copy
```

```python
quarters = quarterly_data_copy['Quarters']
GDP_growth_change = quarterly_data_copy['Economic growth percent change in quarterly real GDP']
# Definition: The percent change in the Gross Domestic Product from the same quarter LAST YEAR using constant prices.
quart_GDP_growth = quarterly_data_copy['Economic growth quarterly']
# Definition: The percent change in the Gross Domestic Product from the PREVIOUS QUARTER using constant prices. 
# Data are seasonally adjusted.
GDP_in_billions = quarterly_data_copy['GDP billion currency units']
```

```python
x_axis = np.arange(len(quarters))
x_axis
```

```python
plt.plot(x_axis, GDP_growth_change, color='b')
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, quarters, rotation='45')
plt.title('Quarterly GDP Growth Change')
plt.xlabel('Quarter')
plt.ylabel('Percent Change')
plt.tight_layout()
plt.savefig('Figures/Quarterly GDP Growth Change.png')
```

```python
plt.plot(x_axis, quart_GDP_growth, color='b')
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, quarters, rotation='45')
plt.title('Quarterly Economic Growth')
plt.xlabel('Quarter')
plt.ylabel('Percent Growth')
plt.tight_layout()
plt.savefig('Figures/Quarterly Economic Growth.png')
plt.hlines(0, 0, 20, alpha=0.25)
```

```python
plt.plot(x_axis, GDP_in_billions, color='b')
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, quarters, rotation='45')
plt.title('GDP in Billions')
plt.xlabel('Quarter')
plt.ylabel('Billions (USD)')
plt.tight_layout()
plt.savefig('Figures/Quarterly GDP in Billions.png')
```
